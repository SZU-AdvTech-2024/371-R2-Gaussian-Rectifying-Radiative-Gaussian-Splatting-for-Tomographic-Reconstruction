```python
import numpy as np
import matplotlib.pyplot as plt

#  vol_gt.npy 
npy_file_path = 'data_generator/synthetic_dataset/volume_gt/0_spine.npy'

#  vol_gt.npy 
volume = np.load(npy_file_path)

# 
print("Volume shape:", volume.shape)
print('vol dtype', volume.dtype)
#  3D 
def visualize_volume_slices(volume, axis=0, indices=None):
    """
     3D 
    
    :
        volume: 3D  (numpy array)
        axis:  (0: , 1: , 2: )
        indices:  ()
    """
    if indices is None:
        indices = [volume.shape[axis] // 4, volume.shape[axis] // 2, 3 * volume.shape[axis] // 4]
    
    plt.figure(figsize=(15, 5))
    for i, idx in enumerate(indices):
        plt.subplot(1, len(indices), i + 1)
        if axis == 0:
            slice_data = volume[idx, :, :]
        elif axis == 1:
            slice_data = volume[:, idx, :]
        elif axis == 2:
            slice_data = volume[:, :, idx]
        else:
            raise ValueError("Axis must be 0, 1, or 2.")
        
        plt.imshow(slice_data, cmap='gray')
        plt.title(f'Slice {idx} along axis {axis}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

#  z 
visualize_volume_slices(volume, axis=0)

#  y 
visualize_volume_slices(volume, axis=1)

#  x 
visualize_volume_slices(volume, axis=2)
```
